```python
#!/usr/bin/env python

import numpy
import scipy.stats
from matplotlib import pyplot

%matplotlib inline
```

```python
fig, ax = pyplot.subplots(1, 1, figsize=(6, 6))
fs1, fs2 = 16, 14
alphas = (0.2, 0.5, 1, 2, 5)
colors = ('blue', 'dodgerblue', 'black', 'darkorange', 'red')
ll = []
for alpha, color in zip(alphas, colors):
    fit = numpy.arange(0, 10, 0.01)
    d = scipy.stats.gamma.pdf(x=fit, a=alpha)
    l = ax.plot(fit, d, '-', c=color, alpha=0.7, linewidth=3)
    ll.append(l)
ax.set_xlabel('Fitness', fontsize=fs1)
ax.set_ylabel('Probability density', fontsize=fs1)
[x.set_fontsize(fs2) for x in ax.xaxis.get_ticklabels()+ax.yaxis.get_ticklabels()]
ax.set_ylim(-0.05, 2.1)
ax.legend(ll, labels=[r'$\alpha=$'+f'{x:.02f}' for x in alphas], fontsize=fs2, frameon=False)
fig.tight_layout()
fig.savefig('hist_fitness_gammas.pdf', dpi=300)
```

```python
nstep = 11
xs = numpy.arange(nstep)
qs = numpy.array([1 - 1/2**n for n in range(nstep)])
fig, ax = pyplot.subplots(1, 1, figsize=(6, 6))
fs1, fs2 = 16, 14
alphas = (0.2, 0.5, 1, 2, 5)
colors = ('blue', 'dodgerblue', 'black', 'darkorange', 'red')
ll = []
for alpha, color in zip(alphas, colors):
    fit = scipy.stats.gamma.ppf(q=qs, a=alpha)
    fit = fit / fit.max()
    l = ax.plot(xs, fit, 'o-', c=color, alpha=0.6, markersize=5, linewidth=2.5)
    ll.append(l)
ax.set_xlabel('Mutation step', fontsize=fs1)
ax.set_ylabel('Fitness', fontsize=fs1)
ax.set_xticks(numpy.arange(11))
[x.set_fontsize(fs2) for x in ax.xaxis.get_ticklabels()+ax.yaxis.get_ticklabels()]
ax.legend(ll, labels=[r'$\alpha=$'+f'{x:.02f}' for x in alphas], fontsize=fs2, frameon=False)
fig.tight_layout()
fig.savefig('trend_trajadapt_gammas.pdf', dpi=300)
```
